# Prueba de `read_maf`
Este notebook prueba el método `read_maf` del módulo `io.py` que genera un objeto `PyMutation`.

Asegúrate de ajustar la ruta al archivo MAF en la siguiente celda.

```
import sys
import os

# Asegúrate de que el directorio raíz del proyecto esté en PYTHONPATH
project_root = os.path.abspath(os.path.join(os.getcwd(), '..', 'src'))
if project_root not in sys.path:
    sys.path.append(project_root)
print('PYTHONPATH configurado para incluir:', project_root)
```

```
# Ruta al archivo MAF
maf_path = os.path.join(
    project_root,
    'pyMut', 'data', 'examples', 'tcga_laml.maf'
)
```

```
from pyMut.io import read_maf

# Leer el archivo MAF y generar el objeto PyMutation
py_mutation = read_maf(maf_path)
# Mostrar el objeto resultante
print(py_mutation)
```
